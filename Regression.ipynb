```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Create 100 random age entries, set seed for reproduction
age = np.random.RandomState(seed=42).randint(low=20, high=70, size=100)

# Create 100 salary entries scaled based on age and adding variance
salary = [xx * 100 + np.random.normal(scale=500.) for xx in age]

# Resphape age and salary data into array of 100 rows, 1 column
age = np.reshape(age, (len(age), 1))
salary = np.reshape(salary, (len(salary), 1))

plt.scatter(x=age, y=salary)
plt.xlabel('Age')
plt.ylabel("Salary")
plt.show()
```

```python
# Create Linear Regression Classifier
reg_line = LinearRegression()

# Fit data into Classifier
reg_line.fit(age, salary)

print ('Salary Prediction Age 50: ', reg_line.predict(50))
print ('R-Squared Score: ', reg_line.score(age, salary))
print ('Intercept: ', reg_line.intercept_)
print ('Slope: ', reg_line.coef_)
```

```python
plt.scatter(x=age, y=salary)
plt.plot(age, reg_line.predict(age), color='red', linewidth=2)
plt.xlabel('Age')
plt.ylabel("Salary")
plt.show()
```
